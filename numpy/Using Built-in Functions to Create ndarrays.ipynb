```python
# We create a 3 x 4 ndarray full of zeros.
import numpy as np
X = np.zeros((3,4))
print(X)
```

```python
X = np.ones((3,2))
print(X)
```

```python
X = np.eye(5)
print(X)
```

```python
X = np.diag([10, 20, 30])
print(X)
```

```python
X = np.arange(4, 10, 2)
print(X)
```

```python
x = np.linspace(0, 25, 10)
print(x)
```

```python
x = np.linspace(0, 25, 10, endpoint = False)
print(x)
```

```python
x = np.arange(30)
print(x)
```

```python
x = np.reshape(x, (5, 6))
print(x)
```

```python
x = np.random.random((3,3))
print(x)
```

```python
x = np.random.randint(4, 15, (3,2))
print(x)
```

```python
x = np.random.normal(0, 0.1, size=(1000,1000))
print(x)
```
